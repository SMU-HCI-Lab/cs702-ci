```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

plt.style.use("seaborn-v0_8")
```

```python
mean = [0, 0]
cov = [[1, 0],
       [0, 1]]
rv_nc = multivariate_normal(mean, cov)  # No correlation
```

```python
# Plot probability densities along x_2 with various x_1 values and sigma_12 values. Here is one example where x_1 = 3 and sigma_12 = 0.8.
x1_value = 3
s_12 = 0.8

# Create a multivariate normal distribution with the given parameters
mean = [0, 0]
cov = [[1, s_12],
       [s_12, 1]]
rv_c = multivariate_normal(mean, cov)

# Plot the probability densities along x_2
x2 = np.linspace(-5, 5, 1000)
x1_values = np.full(x2.shape, x1_value)
x1x2s = np.dstack((x1_values, x2)) 
pdf_given_x1_00 = rv_nc.pdf(x1x2s)
pdf_given_x1_08 = rv_c.pdf(x1x2s)

fig, axs = plt.subplots(1, 2, figsize=(10, 3))
axs[0].plot(x2, pdf_given_x1_00)
axs[0].set_title(r"Probability Density along $x_2$ when $\sigma_{12}=0$")
axs[0].set(xlabel=r"$x_2$", xlim=[-5, 5], yticks=[])

axs[1].plot(x2, pdf_given_x1_08)
axs[1].set_title(r"Probability Density along $x_2$ when $\sigma_{12}=0.8$")
axs[1].set(xlabel=r"$x_2$", xlim=[-5, 5], yticks=[])

plt.tight_layout()
plt.show()
```
