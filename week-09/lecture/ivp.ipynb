```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt


def fitts(t, x, k, r):
    """
    ODE model for Fitts's law: dx/dt = k(r-x)
    """
    dxdt = k * (r - x)
    return dxdt

# Parameters
k = 0.5
r = 10
t_span = (0, 20)
x0 = [0]

# Solve the initial value problem
solution = solve_ivp(
    fun=lambda t, x: fitts(t, x, k, r),
    t_span=t_span,
    y0=x0,
    dense_output=True,
)
```

```python
# Plot the solution
t = np.linspace(t_span[0], t_span[1], 500)
x_numerical = solution.sol(t)[0]

# Analytical solution for comparison
x_analytical = r * (1 - np.exp(-k * t))

# Plot results
plt.figure(figsize=(10, 6))
plt.plot(t, x_numerical, 'b-', label='Numerical Solution')
plt.plot(t, x_analytical, 'r--', label='Analytical Solution')
plt.grid(True, alpha=0.3)
plt.xlabel('Time (t)')
plt.ylabel('x(t)')
plt.legend()
plt.show()
```
