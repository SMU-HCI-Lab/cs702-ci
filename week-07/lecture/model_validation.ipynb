```python
import matplotlib.pyplot as plt
import numpy as np
import numpyro
import numpyro.distributions as dist
from numpyro.infer import MCMC, NUTS
import jax.numpy as jnp
from jax import random
```

```python
def wait_time_model(wait_times):
    """M/M/1 queue"""
    rate = numpyro.sample("rate", dist.Gamma(1.0, 1.0))
    
    numpyro.sample("wait_times", 
                   dist.Exponential(rate=rate),
                   obs=wait_times)
```

```python
# Synthetic data
key = random.PRNGKey(0)
true_rate = 0.5
synthetic_wait_times = np.random.exponential(scale=1/true_rate, size=100)
```

```python
kernel = NUTS(wait_time_model)
mcmc = MCMC(kernel, num_warmup=1000, num_samples=2000)
mcmc.run(key, wait_times=synthetic_wait_times)
mcmc.print_summary()
```

```python
samples = mcmc.get_samples()
plt.figure()
plt.hist(
    synthetic_wait_times, 
    bins=30, 
    density=True, 
    alpha=0.5, 
    label='Observed data'
    )
    
x = np.linspace(0, max(synthetic_wait_times), 100)
rate_mean = np.mean(samples['rate'])
plt.plot(
    x, 
    rate_mean * np.exp(-rate_mean * x), 
    'k--', 
    label=f'rate={rate_mean:.2f}'
    )
    
plt.xlabel('Wait time')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
def weber_fechner_model(stimuli, responses):
    """Weber-Fechner law"""
    k = numpyro.sample("k", dist.LogNormal(0.0, 1.0))
    reference = numpyro.sample("reference", dist.LogNormal(0.0, 1.0))
    sigma = numpyro.sample("sigma", dist.LogNormal(0.0, 1.0))
    
    expected_response = k * jnp.log(stimuli / reference)
    
    numpyro.sample("responses",
                   dist.Normal(expected_response, sigma),
                   obs=responses)
```

```python
stimuli = np.logspace(0, 2, 50)
true_k = 2.0
true_reference = 1.0
true_responses = true_k * np.log(stimuli / true_reference) + np.random.normal(0, 0.1, size=50)
```

```python
kernel = NUTS(weber_fechner_model)
mcmc = MCMC(kernel, num_warmup=1000, num_samples=2000)
mcmc.run(key, stimuli=stimuli, responses=true_responses)
mcmc.print_summary()
```

```python
samples = mcmc.get_samples()
plt.figure()
    
plt.scatter(
    stimuli, 
    true_responses, 
    alpha=0.5, 
    label='Observed data'
    )
    
# Plot fitted curve
x = np.logspace(0, 2, 100)
k_mean = np.mean(samples['k'])
ref_mean = np.mean(samples['reference'])
plt.plot(
    x, 
    k_mean * np.log(x / ref_mean), 
    'k--',
    label=f'k={k_mean:.2f}, ref={ref_mean:.2f}'
    )
    
#plt.xscale('log')
plt.xlabel('Stimulus intensity')
plt.ylabel('Perceived intensity')
plt.legend()
plt.show()
```

```python
import arviz as az
import jax.numpy as jnp
import matplotlib.pyplot as plt
import numpyro
import numpyro.distributions as dist
from jax import random
from numpyro.infer import MCMC, NUTS, Predictive

# Set random seed
RANDOM_SEED = 42
rng_key = random.PRNGKey(RANDOM_SEED)

# Generate synthetic data
N = 100
true_a, true_b = 0.5, 3.0
true_sd = 2.0

# Generate predictor and outcome data
key, subkey1, subkey2 = random.split(key, 3)
predictor = random.normal(subkey1, shape=(N,)) * 6.0 + 2.0  # loc=2, scale=6
true_mu = true_a + true_b * predictor
outcome = random.normal(subkey2, shape=(N,)) * true_sd + true_mu

# Plot raw data
plt.scatter(predictor, outcome)
plt.xlabel("Predictor")
plt.ylabel("Outcome")
plt.show()
```

```python
def linear_model(predictor=None, outcome=None):
    a = numpyro.sample("a", dist.Normal(0.0, 0.5))
    b = numpyro.sample("b", dist.Normal(0.0, 1.0))
    sigma = numpyro.sample("sigma", dist.Exponential(1.0))

    mu = numpyro.deterministic("mu", a + b * predictor)

    with numpyro.plate("data", len(predictor)):
        numpyro.sample("obs", dist.Normal(mu, sigma), obs=outcome)


rng_key, subkey1 = random.split(rng_key, 2)
kernel = NUTS(linear_model)
mcmc = MCMC(kernel, num_warmup=2000, num_samples=1000, num_chains=1)
mcmc.run(subkey1, predictor=predictor, outcome=outcome)
```

```python
posterior_samples = mcmc.get_samples()
predictive = Predictive(linear_model, posterior_samples)
post_pred_samples = predictive(random.PRNGKey(1), predictor=predictor, outcome=None)

idata = az.from_numpyro(
    mcmc,
    posterior_predictive=post_pred_samples,
    coords={"obs_dim": jnp.arange(len(predictor))},
    dims={"obs": ["obs_dim"]}
)

az.plot_ppc(idata, num_pp_samples=100)
plt.show()
```

```python
loo = az.loo(idata)
print(loo)
```

```python
# Evaluation with test data
```

```python
rng_key, subkey1, subkey2 = random.split(rng_key, 3)
new_x = random.normal(subkey1, shape=(N,)) * 6.0 + 2.0
true_mu = true_a + true_b * new_x
true_y = random.normal(subkey2, shape=(N,)) * true_sd + true_mu

predictive = Predictive(linear_model, posterior_samples)
predictions = predictive(rng_key, predictor=new_x, outcome=None)
pred_mean = jnp.mean(predictions['mu'], axis=0)

# Calculate metrics
mae = jnp.mean(jnp.abs(true_y - pred_mean))
rmse = jnp.sqrt(jnp.mean((true_y - pred_mean) ** 2))

print("\nEvaluation Metrics:")
print(f"MAE: {mae:.3f}")
print(f"RMSE: {rmse:.3f}")
# MAE: 19.616
# RMSE: 23.631
```
